```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Date': ['2020-05-08', '2020-05-07', '2020-05-06', '2020-05-05', '2020-05-08', '2020-05-07','2020-05-06', '2020-05-05'],'Data': [51, 82, 63, 14, 505, 1006, 607, 1208]})
```

```python
df_Data = df.groupby(df['Data'])['Data'].agg({'sums':'sum'})
```

```python
plt.plot(df_Data)
for m, n in zip(df_Data.index, df_Data['sums']):
    plt.text(m, n, n, ha='center', va='bottom', fontsize=15)
plt.show()
```
